```python
class Library:
    def __init__(self, books):
        self.books = books
        self.book_status = {book: None for book in self.books}

    def display_books(self):
        print("Available Books: ")
        for book in self.books:
            if self.book_status[book] == None:
                print(book)
        print()

    def lend_book(self, book, borrower):
        if book in self.books:
            if self.book_status[book] == None:
                self.book_status[book] = borrower
                print(f"\n{book} is now borrowed by {borrower}")
            else:
                print("\nThis book is already borrowed by someone else.")
        else:
            print("\nSorry, this book is not available in the library.")
    
    def return_book(self, book, borrower):
        if book in self.books:
            if self.book_status[book] == borrower:
                self.book_status[book] = None
                print(f"\n{book} is returned successfully by {borrower}")
            else:
                print("\nThis book was not borrowed by this person.")
        else:
            print("\nThis book is not available in the library")
    
    def add_book(self, book):
        self.books.append(book)
        self.book_status[book] = None
        print(f"\n{book} has been added to the library")
    

class Customer:
    def request_book(self):
        self.book = input("Please enter the name of the book you want to borrow: ")
        return self.book

    def return_book(self):
        self.book = input("Please enter the name of the book you want to return: ")
        return self.book

    def borrow_name(self):
        self.name = input("Enter your name: ")
        return self.name



library = Library(['B1', 'B2', 'B3', 'B4', 'B5'])
customer = Customer()
    
while True:
    print("\n======================")
    print("LIBRARY MANAGEMENT SYSTEM")
    print("======================")
    print("1. Display all available books")
    print("2. Request a book")
    print("3. Return a book")
    print("4. Add a book")
    print("5. Exit")
    choice = int(input("Enter your choice: "))

    if choice == 1:
        library.display_books()

    elif choice == 2:
        book = customer.request_book()
        borrower = customer.borrow_name()
        library.lend_book(book, borrower)

    elif choice == 3:
        book = customer.return_book()
        borrower = customer.borrow_name()
        library.return_book(book, borrower)

    elif choice == 4:
        book = input("Please enter the name of the book you want to add: ")
        library.add_book(book)

    elif choice == 5:
        print("\nThank you for using the Library Management System. Have a nice day.")
        break

    else:
        print("Invalid choice. Please choose a number from 1-5.")
```
